```python
# ======================
# Install dependencies
# ======================
!pip install wordcloud matplotlib pandas -q

# ======================
# Imports
# ======================
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# ======================
# Load CSV
# ======================
csv_path = "/kaggle/input/hp-all/Final_results.csv"  # update if needed
df = pd.read_csv(csv_path)

# ======================
# Define the 6 model columns to save
# ======================
model_columns = [
    "styled",
]

# ======================
# Generate and save word clouds
# ======================
for col in model_columns:
    # Combine all text for the column
    text = " ".join(df[col].dropna().astype(str).tolist())

    # Create word cloud
    wc = WordCloud(
        width=1000,
        height=600,
        background_color="white",
        colormap="viridis",
        collocations=False,
        max_words=200
    ).generate(text)

    # Save to PNG
    output_filename = f"{col}_wordcloud.png"
    wc.to_file(output_filename)
    print(f"Saved {output_filename}")
```
